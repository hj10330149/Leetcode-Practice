```python
#Time Limit Exceeded

nums=[4,3,2,7,8,2,3,1]
length=len(nums)
newlist=list(range(1,length+1))
ans=[]
for i in newlist:
    if i not in nums:
        ans.append(i)
return(ans)
```

```python
#Time Limit Exceeded

nums=[4,3,2,7,8,2,3,1]
length=len(nums)
newlist=list(range(1,length+1))
nums=list(set(nums)) #去重複
diff=[x for x in newlist if x not in nums]
return(diff)
```
