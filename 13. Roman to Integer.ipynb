```python
s = "III"
walk, run = 0, 1
Ans = 0
Roman = {
    'I' : 1,
    'V' : 5,
    'X' : 10,
    'L' : 50,
    'C' : 100,
    'D' : 500,
    'M' : 1000
}
while run < len(s):
    if Roman[s[walk]] >= Roman[s[run]]:
        Ans += Roman[s[walk]]
        walk += 1
        run += 1
    else:
        Ans += (Roman[s[run]]-Roman[s[walk]])
        walk += 2
        run += 2

if run == len(s):
    Ans += Roman[s[walk]]

print(Ans)
```
