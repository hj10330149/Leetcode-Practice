```python
#Time Limit Exceeded
#[2,3,-2,4]
nums = [-2,0,-1]

start=0
end=0
max_prod=float('-inf')
current_prod=1

while start < len(nums):

  current_prod *= nums[end]
  max_prod=max(max_prod, current_prod)
  end += 1

  if end == len(nums):
    start += 1
    end = start
    current_prod=1
  
  if current_prod == 0 :
    current_prod=1

print(max_prod)
```

```python
#Solution
nums = [2,3,-2,4,-2,-2,-2]
max_prod = [0]*len(nums)
min_prod = [0]*len(nums)
max_prod[0] = nums[0]
min_prod[0] = nums[0]
result = nums[0]
print(max_prod[0], min_prod[0], result)

# Loop through the array and update max_prod, min_prod, and result
for i in range(1, len(nums)):
    if nums[i] >= 0:
        max_prod[i] = max(nums[i], max_prod[i-1]*nums[i])
        min_prod[i] = min(nums[i], min_prod[i-1]*nums[i])
    else:
        max_prod[i] = max(nums[i], min_prod[i-1]*nums[i])
        min_prod[i] = min(nums[i], max_prod[i-1]*nums[i])
    result = max(result, max_prod[i])
    print(max_prod[i], min_prod[i], result)
```
