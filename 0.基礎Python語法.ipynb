groupby( ) 可以對資料進行分組

```python
data = {
    'emp_id': [1, 1, 1, 2, 2],
    'event_day': ['2020-11-28', '2020-11-28', '2020-12-03', '2020-11-28', '2020-12-09'],
    'in_time': [4, 55, 1, 3, 47],
    'out_time': [32, 200, 42, 33, 74]
}

employees = pd.DataFrame(data)

employees['total_time']=employees['out_time']-employees['in_time']
employees=employees.loc[:, ['event_day', 'emp_id', 'total_time']]
group=employees.groupby(['event_day','emp_id']).sum().reset_index()
print(group)
```

```python
import pandas as pd

data = {
    'sell_date': ['2020-05-30', '2020-06-01', '2020-06-02', '2020-05-30', '2020-06-01', '2020-06-02', '2020-05-30'],
    'product': ['Headphone', 'Pencil', 'Mask', 'Basketball', 'Bible', 'Mask', 'T-Shirt']
}

activities = pd.DataFrame(data)

activities=activities.drop_duplicates().sort_values('product')
grouped = activities.groupby('sell_date').agg(num_sold=('product', 'size'), products=('product', ','.join)).reset_index() #groupby後想做兩件事以上可以用agg()
print(grouped)
```

concat( )：利用合併多個Pandas DataFrame的方式來新增資料，並且會回傳一個新的Pandas DataFrame資料集

```python
#concat
grades = {
    "name": ["Mike", "Sherry", "Cindy", "John"],
    "math": [80, 75, 93, 86],
    "chinese": [63, 90, 85, 70]
}

df1 = pd.DataFrame(grades)
print("原來的df")
print(df1)

print("=================================")

df2 = pd.DataFrame({
    "name": ["Henry"],
    "math": [60],
    "chinese": [62]
})

new_df = pd.concat([df1, df2], ignore_index=True)
print("合併df來新增資料")
print(new_df)
```

刪除含有NA的資料

dropna( )：刪除含有NaN或空值的資料，並且會回傳一個新的Pandas DataFrame資料集，在進行資料清理的時候非常實用

drop_duplicates( )：刪除重複的資料，並且會回傳一個新的Pandas DataFrame資料集，同樣最常應用在資料清理的時候，想要針對特定欄位時候可以用 subset=['salary']

https://zhuanlan.zhihu.com/p/116884554


```python
import pandas as pd
import numpy as np

#dropna
grades = {
    "name": ["Mike", "Sherry", np.NaN, "John"],
    "city": ["Taipei", np.NaN, "Kaohsiung", "Taichung"],
    "math": [80, 75, 93, 86],
    "chinese": [63, 90, 85, 70]
}

df = pd.DataFrame(grades)
print("原來的df")
print(df)

print("======================================")

new_df = df.dropna()
print("刪除空值後的df")
print(new_df)

print("============================================================================")

#drop_duplicates
grades = {
    "name": ["Mike", "Mike", "Cindy", "John"],
    "city": ["Taipei", "Taipei", "Kaohsiung", "Taichung"],
    "math": [80, 80, 93, 86],
    "chinese": [80, 80, 93, 86]
}

df = pd.DataFrame(grades)
print("原來的df")
print(df)

print("======================================")

new_df = df.drop_duplicates()
print("刪除重複值後的df")
print(new_df)
```

排序Pandas DataFrame資料

sort_index( )：依照索引值來進行排序，並且會回傳一個新的Pandas DataFrame資料集

sort_values( )：依照欄位內容來進行排序，並且會回傳一個新的Pandas DataFrame資料集

rank(method='dense', ascending=False) 用於在一個 DataFrame 或 Series 中計算數據的排名。

```python
grades = {
    "name": ["Mike", "Sherry", "Cindy", "John"],
    "math": [80, 75, 93, 86],
    "chinese": [63, 90, 85, 70]
}

df = pd.DataFrame(grades)
df.index = ["s3", "s1", "s4", "s2"]  # 自訂資料索引值

print("原來的df")
print(df)

print("============================")

new_df = df.sort_index(ascending=True) #依照index遞增
print("遞增排序")
print(new_df)

print("============================")

new_df = df.sort_index(ascending=False) #依照index遞減
print("遞減排序")
print(new_df)
```

```python
grades = {
    "name": ["Mike", "Sherry", "Cindy", "John"],
    "math": [80, 75, 93, 86],
    "chinese": [63, 90, 85, 70]
}

df = pd.DataFrame(grades)

print("原來的df")
print(df)

print("============================")

new_df = df.sort_values(["math"], ascending=True) #依照math分數大小遞增
print("遞增排序")
print(new_df)

print("============================")

new_df = df.sort_values(["math"], ascending=False) #依照math分數大小遞減
print("遞減排序")
print(new_df)
```

```python
import pandas as pd

# 假設這是您的三個列表
ids = [1, 2, 3, 4, 5, 6]
scores = [3.50, 3.65, 4.00, 3.85, 4.00, 3.65]

# 將列表組合成 DataFrame
data = {'id': ids, 'score': scores}
scores = pd.DataFrame(data)
print(scores)

# Use the rank method to assign ranks to the scores in descending order with no gaps
scores['rank'] = scores['score'].rank(method='dense', ascending=False)

# Drop id column & Sort the DataFrame by score in descending order
result_df = scores.drop('id',axis=1).sort_values(by='score', ascending=False)

print(result_df)
```

用apply新增資料做變數篩選

Python 功能讀取一行（Row）pandas dataframe 的數據，並輸出該行在新列（Column）的值。

```python
import pandas as pd
df = pd.DataFrame(
    {
      'Name':   ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M', 'N', 'O'],
      'Sex':    ['M', 'M', 'M', 'F', 'F', 'M', 'F', 'M', 'F', 'M', 'M', 'M', 'F', 'M', 'M'],
      'Age':    [38, 28, 31, 34, 28, 28, 36, 33, 22, 39, 22, 24, 31, 29, 22],
      'Height': [1.74, 1.51, 1.67, 1.87, 1.8, 1.51, 1.85, 1.89, 1.81, 1.72, 1.75, 1.64, 1.9, 1.62, 1.61],
      'Weight': [45, 63, 39, 45, 67, 66, 53, 45, 72, 46, 58, 44, 73, 70, 51],
    }
)
print(df)

# apply用法
df_eg1 = df.copy()
def BMI_2(weight, height): #先建立一個計算BMI的函數
  return round(weight / (height ** 2), ndigits=2)

#將apply用於計算df_eg1中每一列資料的BMI
df_eg1['BMI_apply2'] = df_eg1.apply(lambda r: BMI_2(r['Weight'],r['Height']), axis=1) #這裡的r代表df_eg1的每一列資料，類似遞迴依序往下
print(df_eg1)
```
